```python
data_sample = data.sample(50000, random_state=42)
```

```python
user_ids = data_sample['id'].tolist()

id_map = {id_: i for i, id_ in enumerate(user_ids)}
```

```python
import pandas as pd

edge_path = "/content/drive/MyDrive/edge.csv"

edges = pd.read_csv(edge_path, chunksize=1000000)
```

```python
edge_list = []

for chunk in edges:
    chunk = chunk[chunk['source_id'].isin(user_ids) & chunk['target_id'].isin(user_ids)]

    for _, row in chunk.iterrows():
        edge_list.append([id_map[row['source_id']], id_map[row['target_id']]])

    if len(edge_list) > 200000:   # limit edges
        break
```

```python
import torch

edge_index = torch.tensor(edge_list, dtype=torch.long).t().contiguous()
```

```python
features = [
    'followers_count',
    'friends_count',
    'statuses_count',
    'verified',
    'protected',
    'desc_length',
    'username_length',
    'ff_ratio',
    'account_age_days'
]

X = data_sample[features].fillna(0).values
X = torch.tensor(X, dtype=torch.float)
```

```python
y = data_sample['label'].map({'human': 0, 'bot': 1}).values
y = torch.tensor(y, dtype=torch.long)
```

```python
from torch_geometric.data import Data

graph_data = Data(x=X, edge_index=edge_index, y=y)
```

```python
import torch.nn.functional as F
from torch_geometric.nn import GCNConv

class BotRGCN(torch.nn.Module):
    def __init__(self):
        super(BotRGCN, self).__init__()
        self.conv1 = GCNConv(X.shape[1], 64)
        self.conv2 = GCNConv(64, 2)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index

        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = self.conv2(x, edge_index)

        return F.log_softmax(x, dim=1)
```

```python
model = BotRGCN()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

for epoch in range(20):
    model.train()
    optimizer.zero_grad()

    out = model(graph_data)
    loss = F.nll_loss(out, graph_data.y)

    loss.backward()
    optimizer.step()

    print(f"Epoch {epoch}, Loss: {loss.item()}")
```

```python
model.eval()
pred = model(graph_data).argmax(dim=1)

from sklearn.metrics import classification_report

print(classification_report(y.numpy(), pred.numpy()))
```

```python
import torch

num_nodes = graph_data.num_nodes

train_mask = torch.zeros(num_nodes, dtype=torch.bool)
test_mask = torch.zeros(num_nodes, dtype=torch.bool)

indices = torch.randperm(num_nodes)

train_size = int(0.8 * num_nodes)

train_mask[indices[:train_size]] = True
test_mask[indices[train_size:]] = True

graph_data.train_mask = train_mask
graph_data.test_mask = test_mask
```

```python
for epoch in range(20):
    model.train()
    optimizer.zero_grad()

    out = model(graph_data)

    loss = F.nll_loss(out[graph_data.train_mask], graph_data.y[graph_data.train_mask])

    loss.backward()
    optimizer.step()

    print(f"Epoch {epoch}, Loss: {loss.item()}")
```

```python
model.eval()

pred = model(graph_data).argmax(dim=1)

from sklearn.metrics import classification_report

print(classification_report(
    graph_data.y[graph_data.test_mask].numpy(),
    pred[graph_data.test_mask].numpy()
))
```

```python
import torch.nn.functional as F
from torch_geometric.nn import GCNConv

class BotRGCN(torch.nn.Module):
    def __init__(self, input_dim):
        super(BotRGCN, self).__init__()

        self.conv1 = GCNConv(input_dim, 128)
        self.conv2 = GCNConv(128, 64)
        self.conv3 = GCNConv(64, 2)

        self.dropout = 0.5

    def forward(self, data):
        x, edge_index = data.x, data.edge_index

        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, p=self.dropout, training=self.training)

        x = self.conv2(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, p=self.dropout, training=self.training)

        x = self.conv3(x, edge_index)

        return F.log_softmax(x, dim=1)
```

# GAT (Graph Attention Network)

```python
import torch
import torch.nn.functional as F
from torch_geometric.nn import GATConv
```

```python
class GAT(torch.nn.Module):
    def __init__(self, input_dim):
        super(GAT, self).__init__()

        self.gat1 = GATConv(input_dim, 64, heads=4, dropout=0.6)
        self.gat2 = GATConv(64 * 4, 32, heads=4, dropout=0.6)
        self.gat3 = GATConv(32 * 4, 2, heads=1, concat=False, dropout=0.6)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index

        x = self.gat1(x, edge_index)
        x = F.elu(x)

        x = self.gat2(x, edge_index)
        x = F.elu(x)

        x = self.gat3(x, edge_index)

        return F.log_softmax(x, dim=1)
```

```python
class_counts = torch.bincount(y)

class_weights = torch.tensor([
    1.0,
    class_counts[0] / class_counts[1]   # moderate boost
])

loss_fn = torch.nn.NLLLoss(weight=class_weights)
```

```python
model = GAT(X.shape[1])
optimizer = torch.optim.Adam(model.parameters(), lr=0.005)
```

```python
for epoch in range(30):
    model.train()
    optimizer.zero_grad()

    out = model(graph_data)

    loss = loss_fn(
        out[graph_data.train_mask],
        graph_data.y[graph_data.train_mask]
    )

    loss.backward()
    optimizer.step()

    print(f"Epoch {epoch}, Loss: {loss.item()}")
```

```python
import torch
from transformers import RobertaTokenizer, RobertaForSequenceClassification
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import classification_report
```

```python
data_sample = data.sample(10000, random_state=42)

texts = data_sample['description'].fillna("").tolist()
labels = data_sample['label'].map({'human': 0, 'bot': 1}).tolist()
```

```python
tokenizer = RobertaTokenizer.from_pretrained('roberta-base')
```

```python
class BotDataset(Dataset):
    def __init__(self, texts, labels, tokenizer):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        encoding = self.tokenizer(
            self.texts[idx],
            padding='max_length',
            truncation=True,
            max_length=32,   # 🔥 reduced
            return_tensors='pt'
        )

        return {
            'input_ids': encoding['input_ids'].squeeze(),
            'attention_mask': encoding['attention_mask'].squeeze(),
            'labels': torch.tensor(self.labels[idx], dtype=torch.long)
        }
```

```python
dataset = BotDataset(texts, labels, tokenizer)

train_size = int(0.8 * len(dataset))
train_dataset, test_dataset = torch.utils.data.random_split(dataset, [train_size, len(dataset)-train_size])

train_loader = DataLoader(train_dataset, batch_size=4, shuffle=True)   # 🔥 small
test_loader = DataLoader(test_dataset, batch_size=4)
```

```python
model = RobertaForSequenceClassification.from_pretrained('roberta-base', num_labels=2)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

print("Using device:", device)
```

```python
from torch.optim import AdamW

optimizer = AdamW(model.parameters(), lr=2e-5)
```

```python
for epoch in range(2):   # 🔥 only 1 epoch
    model.train()

    for i, batch in enumerate(train_loader):
        optimizer.zero_grad()

        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels_batch = batch['labels'].to(device)

        outputs = model(
            input_ids=input_ids,
            attention_mask=attention_mask,
            labels=labels_batch
        )

        loss = outputs.loss
        loss.backward()
        optimizer.step()

        if i % 50 == 0:
            print(f"Step {i}, Loss: {loss.item()}")   # 🔥 keeps session alive
```

```python
model.eval()

preds = []
true_labels = []

with torch.no_grad():
    for batch in test_loader:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        logits = outputs.logits

        preds.extend(torch.argmax(logits, dim=1).cpu().numpy())
        true_labels.extend(batch['labels'].numpy())

print(classification_report(true_labels, preds))
```
